```python
import numpy as np
import pandas as pd
import pickle
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
# Extract year from release_date
movies['year'] = pd.to_datetime(movies['release_date'], errors='coerce').dt.year

# Flatten genres for easy filtering
movies['genres_flat'] = movies['genres'].apply(lambda x: [g['name'] if isinstance(g, dict) else g for g in x] if isinstance(x, list) else [])

# Get unique genres, actors, directors
all_genres = sorted({g for sublist in movies['genres_flat'] for g in sublist})
all_actors = sorted({a for sublist in movies['cast'] for a in sublist})
all_directors = sorted({d for sublist in movies['crew'] for d in sublist})
```

```python

# IMDb weighted rating formula
def weighted_rating(x, m, C):
    v = x['vote_count']
    R = x['vote_average']
    return (v/(v+m) * R) + (m/(m+v) * C)

C = movies['vote_average'].mean()
m = movies['vote_count'].quantile(0.90)
movies['weighted_rating'] = movies.apply(lambda x: weighted_rating(x, m, C), axis=1)
```

```python
import ast

# Extract year from release_date
movies['year'] = pd.to_datetime(movies['release_date'], errors='coerce').dt.year

# Parse genres column to list of names
def parse_genres(obj):
    try:
        return [g['name'] for g in ast.literal_eval(obj)]
    except:
        return []

movies['genres_flat'] = movies['genres'].apply(parse_genres)

# Parse cast column to list of names (limit to top 5 actors)
def parse_cast(obj):
    try:
        return [c['name'] for c in ast.literal_eval(obj)[:5]]
    except:
        return []

movies['cast_flat'] = movies['cast'].apply(parse_cast)

# Parse crew column to list of directors
def parse_director(obj):
    try:
        return [c['name'] for c in ast.literal_eval(obj) if c.get('job') == 'Director']
    except:
        return []

movies['director_flat'] = movies['crew'].apply(parse_director)

# Build unique lists for dropdowns
all_genres = sorted(set(g for sublist in movies['genres_flat'] for g in sublist if g))
all_actors = sorted(set(a for sublist in movies['cast_flat'] for a in sublist if a))
all_directors = sorted(set(d for sublist in movies['director_flat'] for d in sublist if d))

# Save pickles
pickle.dump(all_genres, open('genres.pkl', 'wb'))
pickle.dump(all_actors, open('actors.pkl', 'wb'))
pickle.dump(all_directors, open('directors.pkl', 'wb'))

# Save movies_full.pkl with these new columns
pickle.dump(movies, open('movies_full.pkl', 'wb'))
```

```python

def filter_movies(genre=None, actor=None, director=None, year_range=None, min_rating=None, sort_by=None):
    df = movies.copy()
    if genre:
        df = df[df['genres_flat'].apply(lambda x: genre in x)]
    if actor:
        df = df[df['cast'].apply(lambda x: actor in x)]
    if director:
        df = df[df['crew'].apply(lambda x: director in x)]
    if year_range:
        df = df[(df['year'] >= year_range[0]) & (df['year'] <= year_range[1])]
    if min_rating:
        df = df[df['vote_average'] >= min_rating]
    if sort_by:
        df = df.sort_values(by=sort_by, ascending=False)
    return df

def get_top_movies(n=10, sort_by='weighted_rating'):
    return movies.sort_values(by=sort_by, ascending=False).head(n)[['title', 'year', 'vote_average', 'weighted_rating']]
```

```python
# Modify recommend function in cell 110-123
def recommend(movie, genre=None):
    movie_index = movies[movies['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:]
    recommended_movies = []
    for i in movies_list:
        title = movies.iloc[i[0]].title
        if genre:
            if genre in movies.iloc[i[0]].genres:
                recommended_movies.append(title)
        else:
            recommended_movies.append(title)
        if len(recommended_movies) == 5:
            break
    return recommended_movies
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Split the data and reset index for safe positional access
train_data, test_data = train_test_split(new_data, test_size=0.2, random_state=42)
train_data = train_data.reset_index(drop=True)

# Recompute vectorizer and similarity for train_data only
cv = CountVectorizer(max_features=5000, stop_words='english')
train_vectors = cv.fit_transform(train_data['tags']).toarray()
train_similarity = cosine_similarity(train_vectors)

# You need access to the original genres for each movie
# Merge genres into new_data for lookup
genres_map = movies[['title', 'genres']].set_index('title')['genres'].to_dict()

def recommend(movie):
    if movie not in train_data['title'].values:
        return []
    movie_pos = train_data[train_data['title'] == movie].index[0]
    distances = train_similarity[movie_pos]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]
    recommended_titles = [train_data.iloc[i[0]].title for i in movies_list]
    return recommended_titles

# Evaluate recommendations using genre overlap
total_correct_recommendations = 0
total_possible_recommendations = 0

for movie in test_data['title']:
    recs = recommend(movie)
    if not recs:
        continue
    test_genres = set(genres_map.get(movie, []))
    for rec in recs:
        rec_genres = set(genres_map.get(rec, []))
        # Count as correct if at least one genre overlaps
        if test_genres & rec_genres:
            total_correct_recommendations += 1
        total_possible_recommendations += 1

# Calculate accuracy (fraction of recommendations with genre overlap)
if total_possible_recommendations > 0:
    accuracy = total_correct_recommendations / total_possible_recommendations
else:
    accuracy = 0

print(f"Accuracy (genre overlap): {accuracy:.2f}")
```
